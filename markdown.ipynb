````python
import os

def py_to_markdown(py_file_path):
    md_file_path = os.path.splitext(py_file_path)[0] + ".md"

    with open(py_file_path, 'r', encoding='utf-8') as f:
        code_content = f.read()

    markdown_content = f"```python\n{code_content}\n```"

    with open(md_file_path, 'w', encoding='utf-8') as f:
        f.write(markdown_content)

    print(f"Successfully converted {py_file_path} to {md_file_path}")


# Example Usage
py_to_markdown(r"C:\Users\GIDI\Desktop\APPS\PAv1\new_test.py")
````
